```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X=np.array([[0,0],[0,1],[1,0],[1,1]])
Y=np.array([0,1,1,0])
```

```python
model=Sequential()
```

```python
model.add(Dense(4,input_dim=2,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X,Y,epochs=1000,batch_size=1)
```

```python
loss, accuracy = model.evaluate(X, Y)
print("Loss:", loss)
print("Accuracy:", accuracy)
```

```python
predictions=model.predict(X)
print(predictions)
```

```python
model.save('Simple_nn_model.h5')
```

```python
# Get weights and biases of the first layer
weights_layer1, biases_layer1 = model.layers[0].get_weights()
print("Layer 1 weights:", weights_layer1)
print("Layer 1 biases:", biases_layer1)

# Get weights and biases of the second layer
weights_layer2, biases_layer2 = model.layers[1].get_weights()
print("Layer 2 weights:", weights_layer2)
print("Layer 2 biases:", biases_layer2)
```
